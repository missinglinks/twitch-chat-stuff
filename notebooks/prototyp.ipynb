```python
URL = "https://api.twitch.tv/v5/videos/{video_id}/comments?client_id={client_id}&cursor={curor}"
```

```python
comments = []

while True:
    resp = requests.get(URL.format(video_id=VIDEO_ID, client_id=CLIENT_ID, curor=cursor)) 
    resp = resp.json()
    
    comments += resp["comments"]
    print(len(comments))
    
    if "_next" in resp:
        cursor = resp["_next"]
        print(cursor)
    else:
        break
```
